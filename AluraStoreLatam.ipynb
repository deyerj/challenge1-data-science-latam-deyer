### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.info()
```

#1. Análisis de facturación



```python
total_tienda1 = tienda["Precio"].sum()
total_tienda2 = tienda2["Precio"].sum()
total_tienda3 = tienda3["Precio"].sum()
total_tienda4 = tienda4["Precio"].sum()

print("Tienda 1:", total_tienda1)
print("Tienda 2:", total_tienda2)
print("Tienda 3:", total_tienda3)
print("Tienda 4:", total_tienda4)
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda.groupby("Categoría del Producto").size().sort_values(ascending=False)
ventas_categoria_t2 = tienda2.groupby("Categoría del Producto").size().sort_values(ascending=False)
ventas_categoria_t3 = tienda3.groupby("Categoría del Producto").size().sort_values(ascending=False)
ventas_categoria_t4 = tienda4.groupby("Categoría del Producto").size().sort_values(ascending=False)

print("Tienda 1:\n", ventas_categoria_t1)
print("\nTienda 2:\n", ventas_categoria_t2)
print("\nTienda 3:\n", ventas_categoria_t3)
print("\nTienda 4:\n", ventas_categoria_t4)

```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()

print("Tienda 1:", promedio_t1)
print("Tienda 2:", promedio_t2)
print("Tienda 3:", promedio_t3)
print("Tienda 4:", promedio_t4)
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
conteo_t1 = tienda["Producto"].value_counts()
print("Tienda 1:")
print("  Más vendido:", conteo_t1.idxmax(), "con", conteo_t1.max(), "ventas")
print("  Menos vendido:", conteo_t1.idxmin(), "con", conteo_t1.min(), "ventas\n")

# Tienda 2
conteo_t2 = tienda2["Producto"].value_counts()
print("Tienda 2:")
print("  Más vendido:", conteo_t2.idxmax(), "con", conteo_t2.max(), "ventas")
print("  Menos vendido:", conteo_t2.idxmin(), "con", conteo_t2.min(), "ventas\n")

# Tienda 3
conteo_t3 = tienda3["Producto"].value_counts()
print("Tienda 3:")
print("  Más vendido:", conteo_t3.idxmax(), "con", conteo_t3.max(), "ventas")
print("  Menos vendido:", conteo_t3.idxmin(), "con", conteo_t3.min(), "ventas\n")

# Tienda 4
conteo_t4 = tienda4["Producto"].value_counts()
print("Tienda 4:")
print("  Más vendido:", conteo_t4.idxmax(), "con", conteo_t4.max(), "ventas")
print("  Menos vendido:", conteo_t4.idxmin(), "con", conteo_t4.min(), "ventas\n")
```

# 5. Envío promedio por tienda

```python
valor_promedio_envio_t1 = round(tienda["Costo de envío"].mean(), 2)
valor_promedio_envio_t2 = round(tienda2["Costo de envío"].mean(), 2)
valor_promedio_envio_t3 = round(tienda3["Costo de envío"].mean(), 2)
valor_promedio_envio_t4 = round(tienda4["Costo de envío"].mean(), 2)

print("Tienda 1:", valor_promedio_envio_t1)
print("Tienda 2:", valor_promedio_envio_t2)
print("Tienda 3:", valor_promedio_envio_t3)
print("Tienda 4:", valor_promedio_envio_t4)

```

# Gráficos

```python
import matplotlib.pyplot as plt

# 1. Gráfico de barras: Ingresos totales por tienda

# Calculamos ingresos totales por tienda
ingresos = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,5))
plt.bar(nombres_tiendas, ingresos, color='skyblue')
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos (miles)")
plt.xlabel("Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python

# 2. Gráfico de pastel: Distribución de categorías en Tienda 1
# Contamos ventas por categoría en Tienda 1
ventas_categoria_t1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(ventas_categoria_t1, labels=ventas_categoria_t1.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de Categorías en Tienda 1")
plt.axis('equal')  # Para que el pastel sea un círculo
plt.show()


```

```python

# 3. Gráfico de líneas: Calificación promedio vs Costo de envío promedio por tienda

# Calculamos valores
calificaciones = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

costos_envio = [
    tienda["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

fig, ax1 = plt.subplots(figsize=(8,5))

color = 'tab:blue'
ax1.set_xlabel('Tiendas')
ax1.set_ylabel('Calificación Promedio', color=color)
ax1.plot(nombres_tiendas, calificaciones, marker='o', color=color, label='Calificación Promedio')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 5)  # Calificación en escala 1-5

ax2 = ax1.twinx()  # segundo eje Y
color = 'tab:green'
ax2.set_ylabel('Costo de Envío Promedio', color=color)
ax2.plot(nombres_tiendas, costos_envio, marker='s', color=color, label='Costo de Envío Promedio')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Calificación y Costo de Envío Promedio por Tienda')
fig.tight_layout()
plt.show()
```

# **Informe Final para el Sr. Juan: Análisis y Recomendación de Tiendas para Venta**
## **Introducción**
El objetivo de este análisis es evaluar el desempeño de las cuatro tiendas disponibles y recomendar al Sr. Juan cuál debe vender para optimizar sus resultados comerciales, quedándose con las tres tiendas que mejor desempeño y potencial presenten. Para ello, se consideraron aspectos clave como ingresos totales, ventas por categorías, calificaciones promedio de clientes, productos más y menos vendidos, y costos promedio de envío.

## **Ingresos Totales**
El análisis de ingresos muestra que la Tienda 1 lidera con ingresos de $1,150,880,400.0., seguida por la Tienda 2, 3 y 4 con cifras decrecientes. Esto indica que la Tienda 1 tiene el mayor volumen de ventas y potencial económico.

## **Categorías Más y Menos Vendidas**
Todas las tiendas comparten tendencias similares en categorías más populares, como “Muebles” y “Electrónicos”, y en las menos vendidas como “Instrumentos musicales”. Sin embargo, algunas tiendas muestran mejores resultados en categorías clave que podrían influir en la decisión.

## **Calificaciones Promedio de Clientes**
La Tienda 1 y Tienda 2 obtienen las mejores calificaciones promedio, superando a la Tienda 3 y 4, lo cual refleja mayor satisfacción y confianza de los clientes.

## **Productos Más y Menos Vendidos**
El análisis de productos más vendidos revela que la Tienda 1 tiene una mayor concentración en productos exitosos con alta rotación, mientras que la Tienda 4 tiene más productos con bajas ventas, lo cual puede indicar menor eficiencia en inventarios.

## **Costos Promedio de Envío**
La Tienda 4 presenta el costo promedio de envío más bajo, lo que es positivo para la competitividad, pero no compensa sus bajos ingresos y menor satisfacción del cliente.

## **Conclusión y Recomendación**
Basándonos en el análisis de ingresos, satisfacción del cliente, y rendimiento en ventas, se recomienda que el Sr. Juan venda la Tienda 4, ya que presenta menores ingresos, calificaciones y rendimiento en productos, a pesar de su bajo costo de envío.

Se sugiere mantener las Tiendas 1, 2 y 3, que muestran un mejor equilibrio entre rentabilidad, satisfacción del cliente y oferta de productos.

Esta decisión permitirá al Sr. Juan optimizar sus recursos y enfocarse en las tiendas con mayor potencial para incrementar sus ganancias y fidelizar clientes.
